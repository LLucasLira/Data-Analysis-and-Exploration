## Visualização das Informações Hospitalares

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Limpeza da coluna '2008/Jan' (remover texto ou valores não numéricos)
dados_hospitalares_1['2025/Jan'] = dados_hospitalares_1['2025/Jan'].replace(r'[^\d.-]', '', regex=True)

# 2. Converter a coluna para numérico, forçando qualquer valor inválido a se tornar NaN
dados_hospitalares_1['2025/Jan'] = pd.to_numeric(dados_hospitalares_1['2025/Jan'], errors='coerce')

# 3. Tratar valores 'NaN' (se necessário)
dados_hospitalares_1['2025/Jan'] = dados_hospitalares_1['2025/Jan'].fillna(0)

# 4. Verifique se os dados estão corretos
print(dados_hospitalares_1[['Unidade da Federação', '2025/Jan']].head())  # Verifique se os dados foram corrigidos

# 5. Plotar o gráfico
plt.figure(figsize=(12, 8))  # Aumenta o tamanho do gráfico para acomodar melhor as etiquetas

# Criar gráfico de barras
dados_hospitalares_1.plot(x='Unidade da Federação', y='2025/Jan', kind='bar', color='blue')

# Personalizar o gráfico
plt.title("Valores de 2025/Jan por Unidade da Federação")
plt.xlabel("Unidade da Federação")
plt.ylabel("Valores de 2025/Jan")

# Exibir o gráfico
plt.show()
```

## Desafio 3: Deixar as legendas ânguladas (0, 45 etc) para facilitar a leitura

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Limpeza da coluna '2008/Jan' (remover texto ou valores não numéricos)
# Remover qualquer coisa que não seja número ou ponto (como caracteres especiais ou espaços extras)
dados_hospitalares_1['2008/Jan'] = dados_hospitalares_1['2008/Jan'].replace(r'[^\d.-]', '', regex=True)

# 2. Converter a coluna para numérico, forçando qualquer valor inválido a se tornar NaN
dados_hospitalares_1['2008/Jan'] = pd.to_numeric(dados_hospitalares_1['2008/Jan'], errors='coerce')

# 3. Tratar valores 'NaN' (se necessário)
# Aqui você pode substituir por 0, pela média ou pela mediana
dados_hospitalares_1['2008/Jan'] = dados_hospitalares_1['2008/Jan'].fillna(0)

# 4. Verifique se os dados estão corretos
print(dados_hospitalares_1[['Unidade da Federação', '2008/Jan']].head())  # Verifique se os dados foram corrigidos

# 5. Plotar o gráfico
plt.figure(figsize=(12, 8))  # Aumenta o tamanho do gráfico para acomodar melhor as etiquetas

# Criar gráfico de barras
dados_hospitalares_1.plot(x='Unidade da Federação', y='2008/Jan', kind='bar', color='blue')

# Personalizar o gráfico
plt.title("Valores de 2008/Jan por Unidade da Federação")
plt.xlabel("Unidade da Federação")
plt.ylabel("Valores de 2008/Jan")

# Ajustar o ângulo das etiquetas para 45 graus e garantir o espaço
plt.xticks(rotation=45, ha='right')  # 'ha' para alinhar corretamente as etiquetas
plt.tight_layout()  # Ajuste automático para evitar sobreposição

# Exibir o gráfico
plt.show()
```

## Limpar dados vazios com a média

```python
# Armazenar a coluna de "Unidade da Federação" antes de fazer alterações
unidade_federacao = dados_hospitalares_1["Unidade da Federação"]

# Remover a coluna de "Unidade da Federação" temporariamente para ajustar as colunas numéricas
dados_hospitalares_1 = dados_hospitalares_1.drop(columns=["Unidade da Federação"])

# Converter as colunas restantes para numérico, forçando erro para 'coerce' (transformando não numéricos em NaN)
dados_hospitalares_1 = dados_hospitalares_1.apply(pd.to_numeric, errors='coerce')

# Adicionar a coluna de "Unidade da Federação" de volta
dados_hospitalares_1["Unidade da Federação"] = unidade_federacao

# Verificar os dados ajustados
print(dados_hospitalares_1.head())

# Calcular a média das colunas numéricas
media_dados = dados_hospitalares_1.mean(numeric_only=True)

# Exibir a média
print(media_dados)
```

```python
colunas_usaveis = media_dados.index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
dados_usaveis = dados_hospitalares_1[colunas_usaveis]
dados_usaveis.head()
```

## Desafio 1: Ajustar Legenda

```python
import matplotlib.pyplot as plt

dados_usaveis.T.plot(figsize=(18,9)) 

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=9, fontsize=8, title="Federação", title_fontsize=12, labelspacing=1.5, handletextpad=1)

# Personalizando os ticks
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Ativando o grid
plt.grid(True, linestyle='--', color='gray', alpha=0.7)

# Exibindo o gráfico
plt.show()
```

## Desafio 2: Retocar os Títulos

```python
import matplotlib.pyplot as plt

dados_usaveis.T.plot(figsize=(18,9)) 

plt.style.use('dark_background')
plt.title('Volume por Federação', fontsize=16, color='white')
plt.ylabel('Volumetria', fontsize=12, color='white')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=9, fontsize=8, title="Federação", title_fontsize=12, labelspacing=1.5, handletextpad=1)

# Personalizando os ticks
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Ativando o grid
plt.grid(True, linestyle='--', color='gray', alpha=0.7)

# Exibindo o gráfico
plt.show()
```
